```python
import spacy

# Cargar el modelo de lenguaje en español
nlp = spacy.load("es_core_news_sm")

# Función para procesar y responder a las entradas del usuario
def respond_to_user(input_text):
    # Procesar el texto con spaCy
    doc = nlp(input_text.lower())

    # Buscar palabras clave para determinar una respuesta
    if any(token.text == "hola" for token in doc):
        return "¡Hola! ¿En qué puedo ayudarte?"
    elif any(token.text == "adiós" for token in doc):
        return "¡Hasta luego! Espero haber sido de ayuda."
    elif any(token.text == "nombre" for token in doc):
        return "Me llamo ChatBotSpacy."
    else:
        return "Disculpa, no entiendo lo que me quieres decir."

# Ejecutar el chatbot
def run_chatbot():
    print("ChatBotSpacy: ¡Hola! Soy un chatbot creado con spaCy. Puedes hablar conmigo.")

    # Inicia bucle para comenza conversación en chatbot
    while True:
        user_input = input("Tú: ")
        response = respond_to_user(user_input)
        print("ChatBotSpacy:", response)

        if 'adiós' in user_input.lower():  # Si la frase del usuario incluye la palabra "adiós" el bucle finaliza.
            break

# Iniciar el chatbot
run_chatbot()
```
