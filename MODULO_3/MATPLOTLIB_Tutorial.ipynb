## <u>***1.- Crear una Gráfica Básica***</u>

A continuación importamos las librerías de *Pandas*, *Numpy* y el módulo de *Pyplot* dentro de la librería de *MatplotLib*.

>Pyplot es una interfaz de alto nivel para la biblioteca Matplotlib de Python. Es una herramienta que permite crear gráficos y diagramas de forma sencilla e intuitiva.
>
>Pyplot proporciona una forma sencilla de utilizar Matplotlib. En lugar de tener que escribir código complejo para crear gráficos, Pyplot ofrece una serie de funciones que permiten crear gráficos de forma rápida y sencilla.

La expresión `"%matplotlib inline"` es una directiva del motor de Jupyter Notebook que indica que los gráficos creados con Matplotlib se deben mostrar dentro del cuaderno, en lugar de abrirse en una nueva ventana.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

Podemos crear graficas tan sencillas como esta, donde creamos dos variables `x` e `y` que contienen unas listas de valores, que se usarán para crear el gráfico.

> El método `plot()` traza (plot) `y` frente a `x` como líneas y/o marcadores.
>
> El método `show()` de Pyplot muestra el gráfico creado en la pantalla. Este método es necesario para mostrar cualquier gráfico creado con Pyplot.

Mas info en [Matplotlib](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html![chrome_1310.png](chrome_1310.jpg)).

```python
x= [1, 2, 3, 4, 5]
y= [2, 4, 6, 8, 10]

plt.plot(x, y)

plt.show()
```

```python
# Tambien podemos crear graficas tan impresionantes como ésta

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
X = np.arange(-5, 5, 0.15)
Y = np.arange(-5, 5, 0.15)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='viridis')
plt.show()
```

```python
# O como esta

x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
```

### Gráficos de Círculos o Scatter

A continuación, vamos a ver gráficas de círculos o Scatter

> +   La `x` muestra la antiguedad en años de 13 coches.
> +   La `y` muestra la velocidad a la que van por una carretera.

A diferencia de `plot()`, el método `scatter()` crea gráfico de dispersión de la variable `y` frente a `x` con distintos tamaños y/o colores de marcador.

Mas info en [Matplotlib](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html).

```python
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]


plt.scatter(x, y)

plt.show()

# EL gráfico nos muestra que cuanto más nuevo es el coche, a mas velocidad va.
```

En el siguiente snippet estamos mostrando dos conjuntos de coordenadas en un mismo gráfico gracias al método `scatter()`.

```python
# Dia 1, la edad y velociadad de 13 coches:

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]
plt.scatter(x, y)


# Dia 2, la edad y velocidad de 15 coches:

x = [2,2,8,1,15,8,12,9,7,3,11,4,7,14,12]
y = [100,105,84,105,90,99,90,95,94,100,79,112,91,80,85]
plt.scatter(x, y)

plt.show()

# Por default los colores son azul y naranja.
# El nuevo gráfico nos indica tambien que cuanto mas nuevo es el coche a mas
# velocidad conduce.
```

Podemos cambiar el color de los puntos mediante el parámetro `color = 'nombre_color'`.

```python
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]
plt.scatter(x, y, color='black')


x = [2,2,8,1,15,8,12,9,7,3,11,4,7,14,12]
y = [100,105,84,105,90,99,90,95,94,100,79,112,91,80,85]
plt.scatter(x, y, color='green')

plt.show()
```

Podemos, incluso, darle un color a cada punto del gráfico.

Lo que hacemos es crear una variable que almacena los valores en un array y luego pasamos esa variable como valor del parametro `c=` del método `scatter()`.

( <u>*NOTA*</u>: *En esta ocasión usamos un array de la librería de Numpy para almacenar los valores, en vez de una lista de Python*).

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])

plt.scatter(x, y, c=colors)

plt.show()
```

Otra opción que tenemos es darle un tamaño diferente a los puntos del gráfico.

Para ello, volvemos a crear una variable que alamacena un array con los valores del tamaño de cada punto y le pasamos esa variable como valor, pero esta vez al parámetro `s=`, del método `scatter()`.

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes)

plt.show()
```

Y una opción más es la posibilidad de ajustar la intensidad del color con el argumento alpha (transparencia).

Para ello, utilizamos el parámetro `alpha=`que tendrá un valor float de *`0.0`* a *`1.0`*.

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes, alpha=0.80)

plt.show()
```

### Gráficas de Histograma

Otro tipo de grafico muy común es el *histograma*.

El histograma muestra el número de observaciones dentro de cada intervalo. Muestra distribuciones de frecuencia.

>El método `random()` de **Numpy** se utiliza para generar números aleatorios.

> El método `normal()` de la librería de **NumPy** genera muestras de una distribución normal.
>
> La *distribución normal*, también conocida como <u>*distribución gaussiana*</u>, es una distribución de probabilidad continua que se caracteriza por su forma de campana.
>
>El método `normal()` tiene tres parámetros:
>
> +   `loc`: El valor de la media de la distribución.
> +   `scale`: El valor de la desviación estándar de la distribución.
> +   `size`: El número de muestras a generar.

>El método `hist()` genera un gráfico de histograma. Ver más en [Matplotlib](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html).

Si por ejemplo preguntamos la altura a 250 personas, podriamos obtener un histograma como el siguiente:

```python
x= np.random.normal(170, 10, 250)

# 170 es el valor medio (loc).
# 10 es la desviacion estandar (scale).
# 250 es el numero de observaciones (size).

plt.hist(x)
plt.show()
```

### Gráficas de tarta o Pizza

Estas gráficas utilizan una sola variable para generarlo.

La gráfica se generan empezando por la mitad de la pizza (corte horizontal), y se mueve de derecha a izquierda.

> El método `pie()` genera gráficos de tarta o pizza. Ver más en [Matplotlib](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pie.html).

```python
x= [1, 2, 3, 4, 10]

plt.pie(x)

plt.show()
```

Podemos añadirle etiquetas a los trozos de pizza, utilizando el parámetro `labels=` y como valor se le pasa una lista de *strings* de Python.

```python
x= [1, 2, 3, 4, 10]
etiquetas=['Peras', 'Platanos', 'Naranjas', 'Manzanas', 'Uvas']

plt.pie(x, labels= etiquetas)

plt.show()
```

Existe la posibilidad de destacar uno de los trozos de pizza usando el parametro
`explode=`, y se pasa lista de valores de tipo *float* de Python.

```python
x= [1, 2, 3, 4, 10]
etiquetas=['Peras', 'Platanos', 'Naranjas', 'Manzanas', 'Uvas']
destacar= [0.4, 0, 0.4, 0, 0]

plt.pie(x, labels= etiquetas, explode= destacar)

plt.show()
```

También le podemos añadir sombras a la gráfica usando el parámetro `shadow=`, y el valor será de tipo booleano (`True` o `False`).

```python
x= [1, 2, 3, 4, 10]
etiquetas=['Peras', 'Platanos', 'Naranjas', 'Manzanas', 'Uvas']
destacar= [0.4, 0, 0.4, 0, 0]

plt.pie(x, labels= etiquetas, explode=destacar, shadow=True)

plt.show()

```

Le podemos cambiar el color a cada trozo de pizza. En este caso, utilizaremos el parámetro `colors=` y como valor le pasamos una lista de *strings* de Python.

```python
x= [1, 2, 3, 4, 10]
etiquetas=['Peras', 'Platanos', 'Naranjas', 'Manzanas', 'Uvas']

colores= ['Green', 'Yellow', 'Orange', 'Red', 'Purple']

plt.pie(x, labels= etiquetas, colors=colores)

plt.show()
```

También le podemos añadir un cartelito (leyenda) a la grafica, que ofrece información de la simbología o colores que se usan en el gráfico, utilizando  el módulo de Matplotlib `legend()`.

> El módulo de `legend()` de Matplotlib se encarga de dibujar leyendas asociadas a los ejes y/o figuras. Esta clase tiene muchas propiedades, entre ellas `fontsize` que permite definir el tamaño de la leyenda.

Ver más en [Matplotlib](https://matplotlib.org/stable/api/legend_api.html).

```python
x= [1, 2, 3, 4, 10]
etiquetas=['Peras', 'Platanos', 'Naranjas', 'Manzanas', 'Uvas']


plt.pie(x, labels= etiquetas)

plt.legend(fontsize='xx-small')
plt.show()
```

Otra propiedad más del módulo `lengend()` que nos permite añadirle un encabezado a la leyenda, es `title`.

```python
x= [1, 2, 3, 4, 10]
etiquetas=['Peras', 'Platanos', 'Naranjas', 'Manzanas', 'Uvas']


plt.pie(x, labels= etiquetas)
plt.legend(title='FRUTAS', fontsize='xx-small')

plt.show()
```

## <u>***3.- Diseño de Markers***</u>

Los markers (marcadores) en Matplotlib son símbolos que se utilizan para representar los puntos de datos en un gráfico sobre los que despues se dibuja la linea.

Pueden utilizarse para una variedad de propósitos, como representar datos individuales, grupos de datos o tendencias.

Los marcadores pueden ser una herramienta útil para mejorar la claridad y la estética de los gráficos. Al elegir los marcadores adecuados, puede ayudar a que sus datos sean más fáciles de entender y analizar.

Los markers se definen mediante el atributo `marker=` de las funciones de trazado de Matplotlib.

El atributo `ms=` especifica el tamaño del marcador. En el siguiente ejemplo, el tamaño del marcador es de 10 píxeles.

```python
x= [1, 2, 3, 4, 5]
y= [2, 4, 6, 8, 10]

plt.plot(x, y, marker='|', ms=10)

plt.show()
```

El atributo `mec=` especifica el color del borde del marker. En nuestro ejemplo, el valor "r" es la abreviatura de "red" (rojo).

```python
x= [1, 2, 3, 4, 5]
y= [2, 4, 6, 8, 10]

plt.plot(x, y, marker='*', ms=10, mec='r')

plt.show()
```

El atributo `mfc=` define el color del interior del marker (black, yellow, red, orange, blue, pink, green, etc).

Como hemos visto en el snippet anterior, algunos colores, aceptan poner solo la primera letra de su nombre (abreaviado), no todos.

```python
x= [1, 2, 3, 4, 5]
y= [2, 4, 6, 8, 10]

plt.plot(x, y, marker='*', ms=10, mfc='pink')

plt.show()
```

## <u>***4.- Diseño de Líneas***</u>

Para el diseño de la líneas del gráfico usamos el atributo `linestyle=` o su alias `ls=`.

Los valores de este atributo admiten tanto palabras predefinidas como símbolos específicos. Puedes ver los tipos de líneas que están predefinidas para este atributo en esta [imagen](https://matplotlib.org/stable/_images/sphx_glr_linestyles_001.png).

```python
x= [1, 2, 3, 4, 5]
y= [2, 4, 6, 8, 10]

#plt.plot(x, y, linestyle='dotted')  -----> En esta opción usamos el nombre del atributo completo y un valor predefinido.
plt.plot(x,y, ls=':')       # -----> Y en esta opción usamos el nombre del atributo abreviado y un símbolo específico.

plt.show()
```

```python
# Otro ejemplo de tipo de línea.

x= [1, 2, 3, 4, 5]
y= [2, 4, 6, 8, 10]

plt.plot(x, y, ls='--')

plt.show()
```

El color de línea también es modificable, usando el atributo `color=`.

```python
x= [1, 2, 3, 4, 5]
y= [2, 4, 6, 8, 10]

plt.plot(x, y, color='r')

plt.show()
```

```python
# Modificamos el color a rosa.

x= [1, 2, 3, 4, 5]
y= [2, 4, 6, 8, 10]

plt.plot(x, y, color='pink')

plt.show()
```

El atributo `linewidth=` o su abreviatura `lw=` permite configurar el ancho de la línea en píxeles.

```python
x= [1, 2, 3, 4, 5]
y= [2, 4, 6, 8, 10]

plt.plot(x, y, lw=10)

plt.show()
```

Aunque los graficos se suelen realizar definiendo los valores de ambos ejes, también es posible dibujar una sola línea definiendo solamente los valores del eje `y`.

Por defecto, los valores del eje `x` son (0, 1, 2, 3, 4, 5, ..., n).

```python
y= np.array([5, 10, 15, 20, 25, 30])

plt.plot(y)
```

Por otro lado, es posible dibujar más de una línea generando un trazado para cada conjunto de valores (arrays), con el método de trazado (*plot*) que estemos usando en cada gráfico, en nuestro es el método `plot()`.

```python
y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])


plt.plot(y1)
plt.plot(y2)


plt.show()
```

## <u>***5.- Etiquetas y Títulos***</u>

En esta sección vamos a usar los siguientes métodos:

+   El método `title()`, establece un título para el gráfico. Ver más en [Matplotlib](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.title.html).

+    El método `xlabel()`, establece una etiqueta para el eje `x`. Ver más en [Matplotlib](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.xlabel.html).

+    El método `ylabel()`, establece una etiqueta para el eje `y`. Ver más en [Matplotlib](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.ylabel.html).

```python
x=[1, 3, 5, 8, 10, 12, 15, 18]
y=[0.3, 0.5, 0.7, 1.10, 1.30, 1.50, 1.60, 1.75]

plt.plot(x, y)

plt.title('Crecimiento del ser humano')
plt.xlabel('Edad')
plt.ylabel('Altura')

plt.show()
```

Podemos cambiarle el tipo de letra (font) a los titulos y etiquetas, utilizando el parámetro `fontdict` en cualquiera de los tres métodos de etiquetado de gráficos.

En este ejemplo hemos creado dos diccionarios `font1` y `font2` que luego pasamos al parámetro `fontdict`, de los métodos `title()` y `xlabel()` e `ylabel` respectivamente.

Los atributos que admite este parámetro se pueden consultar en [Matplotlib](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.title.html).

```python
font_algerian = {'fontname': 'Algerian'}
x=[1, 3, 5, 8, 10, 12, 15, 18]
y=[0.3, 0.5, 0.7, 1.10, 1.30, 1.50, 1.60, 1.75]


font1= {'fontfamily':'fantasy', 'color':'green', 'size':15}
font2= {'fontfamily':'serif','color':'purple', 'size':10}
plt.plot(x, y)

plt.title('Crecimiento del ser humano', fontdict=font1)
plt.xlabel('Edad', fontdict=font2)
plt.ylabel('Altura', fontdict=font2)

plt.show()
```

Podemos cambiar la alineación del texto del título en el gráfico, a la izquierda, derecha o por defecto en el centro. Para ello usamos el parámetro `loc` y los valores `right` y `left`.

```python
x=[1, 3, 5, 8, 10, 12, 15, 18]
y=[0.3, 0.5, 0.7, 1.10, 1.30, 1.50, 1.60, 1.75]


font1= {'family':'serif', 'color':'green', 'size':15}
font2= {'family':'serif', 'color':'purple', 'size':10}
plt.plot(x, y)

plt.title('Crecimiento del ser humano', fontdict=font1, loc='left')
plt.xlabel('Edad', fontdict=font2)
plt.ylabel('Altura', fontdict=font2)

plt.show()
```

Otra opción más que podemos añadir a los gráficos es mostrar las cuadrículas de intersección de los valores `x` e `y`, añadiendo el método `grid()`.

## <u>***6.- Mostrar Varias Gráficas a la Vez***</u>

Vamos a usar el método `subplot()` para mostrar varias gráficas a la vez en pantalla.

El método `subplot()` de Matplotlib crea una subfigura en una figura existente. Una subfigura es una parte de una figura que puede contener un gráfico o un diagrama.

El método `subplot()` tiene los siguientes parámetros:

+   `nrows` (opcional): El número de filas de la subfigura.
+   `ncols` (opcional): El número de columnas de la subfigura.
+   `index` (opcional): El índice de la subfigura.

El índice (`index`) de la subfigura se utiliza para identificar la subfigura. El índice se puede utilizar para acceder a una subfigura en cuestión o para realizar operaciones en la subfigura.

```python
# Plot 1
x = [5, 10, 15, 20]
y = [-4, -8, -14, -30]

plt.subplot(1, 2, 1)
plt.plot(x,y)

# Plot 2:
x = [5, 10, 15, 20]
y = [8, 16, 24, 32]

plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```

Para ver como funciona los paraámetros del método `subplot()`, vamos a cambiar los dos primeros parametros, y los vamos a invertir su valor (`1`, `2`), siendo ahora (`2`, `1`) respectivamente.

Como podemos ver, viendo los gráficos que se han creado, en el snippet anterior se mostró una figura con 1 fila y 2 columnas (un gráfico (plot) en cada columna), mientras que en la figura que se crea en este snippet, se muestra 2 filas (un gráfico en cada fila) y 1 columna.

```python
# Plot 1
x = [5, 10, 15, 20]
y = [-4, -8, -14, -30]

plt.subplot(2, 1, 1)
plt.plot(x,y)

# Plot 2:
x = [5, 10, 15, 20]
y = [8, 16, 24, 32]

plt.subplot(2, 1, 2)
plt.plot(x,y)

plt.show()
```

Vamos a ponerle título a cada gráfico (plot), usando el método `title()`.

```python
# Plot 1:
x = [3, 6, 9, 12]
y = [-4, -8, -14, -30]

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title('TEMPERATURA INVIERNO')


# Plot 2:
x = [3, 6, 9, 12]
y = [8, 16, 24, 32]

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title('TEMPERATURA VERANO')

plt.show()
```

Se le puede añadir un titulo a todos los subplots usando el método `suptitle()`.

> El método `suptitle()` añade un subtítulo centrado al conjunto de las figuras del gráfico.

Más info en [Matplotlib](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.suptitle.html).

```python
#plot 1:
x = [3, 6, 9, 12]
y = [-4, -8, -14, -30]

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title('TEMPERATURA INVIERNO')

#plot 2:
x = [3, 6, 9, 12]
y = [8, 16, 24, 32]

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title('TEMPERATURA VERANO')

plt.suptitle('RUSIA')
plt.show()
```
